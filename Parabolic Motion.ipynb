```python
#Lets have matplotlib "inline"
%pylab inline

#Lets have large and high-res figures
%config InlineBackend.figure_format = 'retina'
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

#Import packages we need
import numpy as np
```

```python
#The analytic solution to this problem
def analytic(t, a, p0, v0):
    return 0.5*a*t**2 + t*v0 + p0
```

```python
nt = 3000

v = np.zeros((nt, 2))
p = np.zeros((nt, 2))
pa = np.zeros((nt, 2))

v[0,:] = np.array([200.0, 100.0])
p[0,:] = np.array([0.0, 0.0])
pa[0,:] = p[0,:]

a = np.array([0.0, -9.81])
dt = 0.1

for n in range(1,nt):    
    t = n*dt
    v[n,:] = v[n-1,:] + dt*a
    p[n,:] = p[n-1,:] + dt*v[n-1,:]
    pa[n,:] = analytic(t, a, p[0,:], v[0,:])
    
    if (p[n,1] < 0.0):
        print "Hit ground at ", str(p[n,:])
        v.resize((n,2))
        p.resize((n,2))
        pa.resize((n,2))
        break;
    
figure()
plot(p[:,0], p[:,1], '-b', label='Simulation')
plot(pa[:,0], pa[:,1], '-r', label='Analytic')
legend()
```
