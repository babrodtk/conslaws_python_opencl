```python
#Lets have matplotlib "inline"
%pylab inline

#Lets have large and high-res figures
%config InlineBackend.figure_format = 'retina'
pylab.rcParams['figure.figsize'] = (10.0, 8.0)

#Import packages we need
import numpy as np
```

```python
#The analytic solution to this problem
def analytic(t, a, p0, v0):
    return 0.5*a*t**2 + t*v0 + p0
```

```python
nt = 3000

#Reserve memory
v = np.empty((nt, 2))
p = np.empty((nt, 2))
prk = np.empty((nt, 2))
pa = np.empty((nt, 2))

#Set initial conditions
v[0,:] = np.array([200.0, 100.0])
p[0,:] = np.array([0.0, 0.0]) #Forward Euler
prk[0,:] = np.array([0.0, 0.0]) #Runge-Kutta 2
pa[0,:] = p[0,:]
a = np.array([0.0, -9.81])
dt = 1

#Simulate
for n in range(1,nt):
    #Calculate t
    t = n*dt
    
    #Calculate velocity and update position
    v[n,:] = v[n-1,:] + dt*a
    
    #Forward Euler
    p[n,:] = p[n-1,:] + dt*v[n-1,:]
    
    #Runge-Kutta 2
    px = prk[n-1,:] + dt*v[n-1,:]
    py = px + dt*v[n,:]
    prk[n,:] = 0.5*(prk[n-1]+py)
    
    
    #Calculate analytic solution
    pa[n,:] = analytic(t, a, p[0,:], v[0,:])
    
    #Stop if we hit the ground
    if (p[n,1] < 0.0):
        print "Hit ground at ", str(p[n,:])
        v.resize((n,2))
        p.resize((n,2))
        prk.resize((n,2))
        pa.resize((n,2))
        break;

#Plot solutions
figure()
plot(p[:,0], p[:,1], 'b.-', label='Simulation (Euler)')
plot(prk[:,0], prk[:,1], 'k--', label='Simulation (RK2)')
plot(pa[:,0], pa[:,1], 'r:', label='Analytic')
legend()
title("Projectile trajectory (dt=" + str(dt) + ")")
axis('tight');
```

```python
v0 = np.array([200.0, 100.0])
p0 = np.array([0.0, 0.0])
dt = 0.1
nt = 100
a = np.array([0.0, -9.81])

for n in range(nt):
    t = float(n)*dt
    v1 = v0+t*a
    p1 = p0 + dt*v1
    
    #Plot
    plot(p1[0], p1[1], 'x')
    
    #Swap p0, p1
    p0, p1 = p1, p0
```
